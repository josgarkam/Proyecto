```
import sys
from wordcloud import WordCloud#, STOPWORDS
from os import path
from PIL import Image
import numpy as np

id = "180118"
print("Bienvenido a WordCloud")
print("Antes de comenzar asegurece de cargar el archivo nubeDePalabras.json")
print("El archivo debe ser cargado en el ide que esté implementando")
print("ingrese el ID del Usuario de Stark Overflow por defecto: ")

#ingreso de ID (Candid Moe) por teclado
i=input()


#Condicional para comparar si el ID es correcto
if i == id :
  
    print("Se comenso a generar la WordCloud")
    print("Por favor espere un momento")
    game =path.dirname("nubeDePalabras.json")

    #Lectura del y escritura del archivo de datos
    with open ("nubeDePalabras.json") as file:#,encoding= "utf-8"
        text = file.read()

    #Impresión de la nube de datos con parametros definidos en una imagen .png
    wc = WordCloud(background_color="black",
		max_words=200,
		height=1600, width=1720,)
    wc.generate(text)
    wc.to_file(path.join(game,"wordcloud.png"))

    # the matplotlib way:
    import matplotlib.pyplot as plt
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")

    # lower max_font_size
    wc = WordCloud(max_font_size=100).generate(text)
    plt.figure()
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()

    print("La WordCloud se generó con exito")
    print("Gracias por utilizar WordCloud")
#Else para pedir ingresar de nuevo el ID 
else:
   print("El ID está mal escrito")
   print("por favor")
   print("Intente de nuevo")
```
